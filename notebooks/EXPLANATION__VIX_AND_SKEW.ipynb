The **SKEW index**, officially known as the **CBOE SKEW Index (ticker: SKEW)**, is a volatility-based indicator published by the **Chicago Board Options Exchange (CBOE)**. It was introduced in 2011 to complement the more widely known **CBOE Volatility Index (VIX)** by providing insight into the perceived **tail risk**‚Äîspecifically, the risk of a large, sudden market decline (often referred to as a ‚Äúblack swan‚Äù event)‚Äîin the **S&P 500 index** over the next 30 days.

---

### **1. What Does the SKEW Index Measure?**

The SKEW index quantifies the **perceived steepness (or asymmetry)** of the implied volatility curve for out-of-the-money (OTM) S&P 500 options. In simple terms:

- **VIX** measures the **expected magnitude** of market volatility (standard deviation of returns) over the next 30 days.
- **SKEW** measures the **expected probability of a large negative return** (typically defined as a move **2 or more standard deviations below the mean**)‚Äîi.e., the risk of extreme downside events.

Whereas the **normal distribution** assumes symmetry in returns (equal likelihood of large up or down moves), financial markets often exhibit **negative skewness**: large downside moves are more feared (and priced) than large upside moves. SKEW captures this market-implied asymmetry.

---

### **2. How Is SKEW Calculated?**

The SKEW index is derived from the **prices of S&P 500 options**‚Äîspecifically, the **implied volatilities of OTM put and call options** across a range of strike prices. The methodology involves:

- Estimating the **risk-neutral probability distribution** of S&P 500 returns over a 30-day horizon using the same framework as the VIX (i.e., model-free, using option prices across strikes).
- From this distribution, calculating the **third moment (skewness)** of the return distribution‚Äîthis is a statistical measure of asymmetry.
- Converting that skewness into an index value using the formula:

  \[
  \text{SKEW} = 100 - 10 \times \text{(Implied Skewness)}
  \]

Thus:
- A **SKEW value of 100** implies a **symmetric return distribution** (i.e., no tail risk beyond what a normal distribution predicts).
- A **SKEW value above 100** (e.g., 130‚Äì145, typical range) implies **negative skewness** and elevated perceived risk of a large downside move.
- **Higher SKEW = greater perceived tail risk.**

> **Typical Range**: SKEW usually trades between **115 and 155**, though it can spike sharply during market stress (e.g., >160 during crises).

---

### **3. Relationship Between SKEW and VIX**

| Feature | **VIX** | **SKEW** |
|--------|--------|---------|
| **What it measures** | Expected **magnitude** of 30-day volatility (standard deviation) | Expected **asymmetry/tail risk** of 30-day returns |
| **Underlying assumption** | How much the market will move | Whether large moves are more likely to be **downside** |
| **Distribution moment** | Second moment (variance) | Third moment (skewness) |
| **Interpretation** | High VIX = high expected volatility | High SKEW = high fear of crash-like events |
| **Response to market stress** | Spikes during volatility surges | Spikes when OTM puts become expensive relative to calls (flight to safety) |

**Key distinction**:  
- You can have **high VIX + low SKEW** ‚Üí high volatility expected, but symmetric (e.g., during a volatile rally).  
- You can have **low VIX + high SKEW** ‚Üí calm markets, but investors hedging against rare crash scenarios (e.g., late-stage bull markets).

---

### **4. Practical Use of SKEW**

- **Risk management**: Portfolio managers use SKEW to gauge demand for **crash protection** (e.g., via OTM puts).
- **Options pricing**: High SKEW implies steeper **volatility skew** (OTM puts more expensive than OTM calls).
- **Market sentiment**: Rising SKEW may signal growing unease about systemic risk, even if VIX is low.

However, **SKEW is less tradable** than VIX:
- There is **no direct ETF or ETN** tracking SKEW (unlike VIX, which has VXX, UVXY, etc.).
- Trading SKEW exposure typically requires **synthetic strategies** using S&P 500 options or variance swaps.

---

### **5. Limitations**

- **Not a forecast**: SKEW reflects current options pricing, not a prediction.
- **Model-dependent**: Though model-free in spirit, real-world bid-ask spreads and liquidity affect accuracy.
- **Less intuitive**: Many investors find VIX easier to interpret than SKEW.
- **Low correlation with realized tail events**: High SKEW doesn't guarantee a crash‚Äîit measures **implied perception**, not inevitability.

---

### **6. Example Scenario**

- **August 2015**: China devaluation scare ‚Üí VIX spiked to ~30, SKEW jumped to ~150.  
- **Late 2019**: Calm markets (VIX ~12), but SKEW rose to ~145 as investors bought cheap tail-risk protection.  
- **March 2020**: VIX exploded to 80+, SKEW initially spiked then **fell**‚Äîbecause options market priced in **broad, extreme volatility**, not just downside skew (everyone was hedging everything).

---

### **Summary**

- **VIX** = ‚ÄúHow much will the market move?‚Äù  
- **SKEW** = ‚ÄúHow likely is a **big drop** (2+ sigma event)?‚Äù  
- Together, they offer a more complete picture of **market fear**: VIX captures general anxiety, SKEW captures **asymmetric crash fear**.

Both are derived from S&P 500 option prices, but they measure **different statistical properties** of the implied return distribution. While VIX is widely cited as the ‚Äúfear gauge,‚Äù SKEW is the ‚Äúblack swan gauge.‚Äù

# üìå 1. **VIX ‚Äì 30-day Implied Volatility of the S&P 500**

### What it is

The **VIX** measures the market‚Äôs expectation of **30-day forward volatility** of the S&P 500 using SPX option prices (mainly out-of-the-money calls and puts).
It is a **fear index** because demand for protection pushes implied volatility up.

### What it tells you

* **Low VIX (10‚Äì14)** ‚Üí complacency, low fear, often late-cycle calm
* **Normal VIX (15‚Äì20)** ‚Üí balanced market
* **High VIX (20‚Äì30)** ‚Üí stress, uncertainty
* **Very high VIX (>30)** ‚Üí crisis or panic

### Regime information

* **Trend followers look at VIX > 25** as a risk-off environment
* **Buy-the-dip works best when VIX spikes above ~30** (historically mean-reverting)

---

# üìå 2. **VIX3M ‚Äì 3-Month Implied Volatility of the S&P 500**

### What it is

**VIX3M** is like the VIX but for **3-month** implied volatility.
Because longer-dated volatility moves more slowly, VIX3M:

* Is **more stable**
* Reflects **macro/structural risk**, not short-term panic
* Moves when institutions hedge long-term exposure

### How it interacts with VIX

The **VIX / VIX3M ratio** is extremely informative:

| Regime              | VIX vs VIX3M                |
| ------------------- | --------------------------- |
| **Normal**          | VIX < VIX3M (contango)      |
| **Stress building** | VIX approaches VIX3M        |
| **Panic / crash**   | VIX > VIX3M (backwardation) |

### Rule of thumb

* **Backwardation (VIX > VIX3M)** = market pricing near-term risk ‚Üí **risk-off**
* **Steep contango (VIX << VIX3M)** = calm ‚Üí **risk-on**

---

# üìå 3. **VVIX ‚Äì Volatility of the VIX**

### What it is

The **VVIX** is the ‚ÄúVIX of the VIX‚Äù: it measures **implied volatility on options on VIX**.

### What it tells you

VVIX reflects how expensive volatility-of-volatility is ‚Äî essentially the **market‚Äôs expectation of VIX turbulence**:

* **Low VVIX (<80)** ‚Üí no one expects volatility shocks
* **High VVIX (>110)** ‚Üí markets expect violent VIX swings
* **Extreme VVIX (>130‚Äì150)** ‚Üí often precedes major selloffs (volatility becomes *reflexive*)

### Why it matters

VVIX rises **before** VIX in many cases because the options market starts pricing **future VIX jumps**.

Think:
**VVIX = fear of future fear**

---

# üìå 4. **SKEW ‚Äì Tail-Risk Index (Left-Tail Probability)**

### What it is

**SKEW** measures demand for **deep out-of-the-money SPX puts**.
It captures **crash risk pricing**, not volatility.

* **High SKEW (>135)** ‚Üí market pricing higher crash probability
* **Low SKEW (<120)** ‚Üí tail risks ignored or cheap

### Why it is different from VIX

* VIX ‚âà pricing of common shocks (gaussian vol)
* SKEW = pricing of *rare* shocks (far-tail events)

You can have:

* **High VIX + low SKEW** ‚Üí panic now, but not expecting further crash
* **Low VIX + high SKEW** ‚Üí calm market but hedgers fear tail risks

[
\text{RegimeScore} = w_1 \cdot TS + w_2 \cdot VolPressure + w_3 \cdot \text{SKEW}
